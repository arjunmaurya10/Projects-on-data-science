GDP Growth of India - EDA

Importing Libraries

```python
import seaborn as sns
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

Getting the data

```python
data = pd.read_csv('India_GDP_Data.csv')
```

```python
plt.figure(figsize = (25, 5))
sns.set_style('dark')
sns.lineplot(data=data.iloc[:, 1:])
```

```python
data.iloc[:,1:].hist(figsize=(20, 4), layout=(1, 4))
```

```python
data.iloc[:,1:].plot(kind = "kde", subplots = True, layout = (1, 4), figsize = (20, 4))

```

```python
sns.pairplot(data, hue='GDP_In_Billion_USD', height=2)
```
